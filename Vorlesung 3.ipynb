surjektiv:

bijektiv: $ f: A \rightarrow B \leftrightarrow f$ injektiv und surjektiv

## Definition III.2:

Eine Menge $A$ heißt abzählbar $:\Leftrightarrow$

- (a) $A$ ist endlich, d.h. $\lvert A\rvert \in \mathbb{N} (\Leftrightarrow \lvert A\rvert < \infty)$, oder
- (b) $A$ ist unendlich, d.h. $\lvert A\rvert = \infty$ und $A$ gleichmächtig zu $\mathbb{N}$

Ist $A$ nicht abzählbar, so heißt $A$ überabzählbar.

# III.2 Dezimaldarstellung von Zahlen

## Definition III.6:

Eine Folge (in $A$) ist eine Abbildung $a:\mathbb{N} \rightarrow A, n \mapsto a_n$

wobei $A \neq \emptyset$. Statt (II.11) schreibt man meist $(a_n)_{n=1}^{\infty}$.

Die Menge aller Folgen in $A$ bezeichnet man mit $A^{\mathbb{N}} := \{\mathbb{N} \rightarrow A\}$.

## Definition III.7:

Eine Folge $a: \mathbb{N} \rightarrow \{0,1,2, \dots, 9\}$ heißt Dezimalsdarstellung der Zahl $0,a_1 a_2 a_3\dots$

(III.12) $$:\Leftrightarrow \forall n\in\mathbb{N} \quad \exists k \in \mathbb{N}, k>n: a_k < 9$$

Die Menge der Dezimaldarstellungen bezeichnen wir mit $\mathbb{D}$.

Die Bedingung (II.12) schließt die Periode 9 aus, wie etwa $$ x = 0,17299999\dots$$

weil $x = 0,1730000\dots$ und Dezimaldarstellung eindeutig sein soll.

## Lemma III.8

Die Menge der $\mathbb{D}$ der Dezimaldarstellungen und die Menge $[0,1) = \{x \in \mathbb{R} \vert 0 \geq x < 1\}$ sind gleichmächtig.

Beweisidee:

Sei $(a_k)_{k=1}^\infty\in\mathbb{D}$ eine Dezimaldarstellung. Dann bilden wir eine Folge $(q_n)_{n=1}^\infty\in\mathbb{Q}^{+^\mathbb{N}}$ rationaler Zahlen durch

- $q_1 := 0,a_1 = a_1\cdot 10^{-1}$

- $q_2 := 0,a_1a_2 = a_1\cdot 10^{-1} + a_2\cdot 10^{-2} \Rightarrow q_1 \le q_2 \le q_1\cdot 10^{-1}$

- $q_3 := 0,a_1a_2a_3 = a_1\cdot 10^{-1} + a_2\cdot 10^{-2} + a_3\cdot 10^{-3} \Rightarrow q_2 \le q_3 \le q_2\cdot 10^{-2}$

$$\vdots$$

Dann bilden wir die Menge $A = \{q_1, q_2, q_3, \dots\} \subseteq \mathbb{Q} \subseteq \mathbb{R}$

Offensichtlich ist $A$ nach durch $1$ beschränkt.

Nach dem Supremums-axiom besitzt $A$ ein (eindeutiges) Supremum, $sup(A) \in [0,1]$

Die gesuchte Bijektion $J$ zwischen $\mathbb{D}$ und $[0,1)$ ergibt sich durch $J[(a_k)_{k=1}^\infty := sup(A)]$.

## Satz III.9

$\mathbb{D}$ ist überabzählbar

(Folgerung: Satz III.10: $\mathbb{R}$ ist überabzählbar.)

Beweis: Nehmen wir an $\mathbb{D}$ wäre abzählbar und 

$$ \mathbb{D} =\{
(a_n^{(1)})_{n=1}^\infty,
(a_n^{(2)})_{n=1}^\infty,
(a_n^{(3)})_{n=1}^\infty,
\dots \} $$

wäre eine Abzählung. 

Definiere nun $(b_n)_{n=1}^\infty\in\{0,1,\dots,9\}^\mathbb{N}$ durch

$$
b_n := \begin{cases}
1 & falls \; a_n^{(n)} \in\{5,6,7,8,9\} \\
7 & falls \; a_n^{(n)} \in\{0,1,2,3,4\} 
\end{cases}
$$

Weil die Ziffer "9" in $(b_n)^\infty_{n=1}$ nie vorkommt, ist $(b_n)^\infty_{n=1}$ eine Dezimaldarstellung, $(b_n)^\infty_{n=1} \in\mathbb{D}$.

Andererseits ist jedoch $a_k^{n} \neq b_n$, daher $(b_k)_{k=1}^\infty \neq (a_k^{(n)})_{k=1}^\infty $. Dies ... alle $n\in\mathbb{N}$, also ist $(b_k)_{k=1}^\infty \notin\mathbb{D}$.

Widerspruch.

$$Z^{(2)} = 0, a_1^{(2)}, a_2^{(2)}, a_3^{(2)}, a_4^{(2)} \dots$$

$$Z^{(3)} = 0, a_1^{(3)}, a_2^{(3)}, a_3^{(3)}, a_4^{(3)} \dots$$
$$\vdots$$


$$\mathbb{D} = \{Z^{(1)}, Z^{(2)}, Z^{(3)}, \dots\}$$
