## VII.2 Minima, Maxima und Mittelwertsatz

### Definition VII.6 

Seien $X \subseteq \mathbb{K}, X \neq \emptyset,$ und $f: x \rightarrow \mathbb{R}$

- (i) $x_0 \in X$ heißt lokales Minimum von $f :\Leftrightarrow \exists\delta > 0 \;\forall x \in B_\mathbb{K}(x_0, \delta) \cap X: \quad f(x) \geq f(x_0)$ ,

- (ii) $x_0$ ist ein globales Minimum von $f: \forall x \in X: f(x) \geq f(x_0)$

- (iii) $x_0 \in X$ heißt lokales Maximum von $f :\Leftrightarrow \exists\delta > 0 \;\forall x \in B_\mathbb{K}(x_0, \delta) \cap X: \quad f(x) \leq f(x_0)$

## Satz VII.7

Seien $X \subseteq \mathbb{R}, X \neq \emptyset, x_0 \in X$ ein Innerer Punkt von $X$ und $f$ bei $x_0$ diff'bar. Dann gilt:

Aus der Schule Bekannt:

$x_0$ ist ein lokales Minimum $\Rightarrow f^\prime(x_0) = 0$,

$x_0$ ist ein lokales Maximum $\Rightarrow f^\prime(x_0) = 0$.

Beweis:

Weil $x_0$ ein Innerer Punkt von $X$ ist, gibt es ein $\delta > 0$ so, dass $(x_0 - \delta, x_0 + \delta) \subseteq X$ und $f(x) \geq f(x_0), \forall x \in (x_0 - \delta, x_0 + \delta)$.

Wir definieren $x^{±}_n := x_0 ± \frac{\delta}{2n} \in (x_0 - \delta, x_0 + \delta), \quad x^{±}_n \neq x_0, \;\forall n \in \mathbb{N}$, und $x^{±}_n \longrightarrow x_0, n \rightarrow \infty$.

Somit ist 
$$f^\prime(x_0) = \lim\limits_{n \longrightarrow \infty} \frac{f(x^{±}_n - f(x_0)}{x^{±}_n - x_0} = \lim\limits_{n \longrightarrow \infty} \underbrace{\frac{2n}{\delta}}_{> 0} \underbrace{[f(x^{±}_n - f(x_0)]}_{\geq 0} \geq 0$$
und andererseits
$$f^\prime(x_0) = \lim\limits_{n \longrightarrow \infty} \frac{f(x^{±}_n - f(x_0)}{x^{±}_n - x_0} = \lim\limits_{n \longrightarrow \infty} \underbrace{\frac{2n}{\delta}}_{< 0} \underbrace{[f(x^{±}_n - f(x_0)]}_{\leq 0}$$

also muss $f^\prime(x_0) = 0$ gelten.

## Satz VII.8 Mittelwertsatz der Differenzialrung

Seien $a,b \in \mathbb{R}, a < b, f: [a,b] \rightarrow \mathbb{R}$ stetig in $[a,b]$ und diff'bar auf $(a,b)$. Dann gibt es einen Punkt $x_0 \in (a,b)$ so, dass 

$$\frac{f(b) - f(a)}{b-a} = f^\prime(x_0).$$

Beweis:

Wir definieren $g: [a,b] \rightarrow \mathbb{R}$,

$$\forall x \in [a,b]:\quad g(x) := \frac{f(b) - f(a)}{b-a} \cdot x - f(x).$$

Damit ist auch $g$ stetig in $[a,b]$ und diff'bar auf $(a,b)$.

1. Fall $g(x) = c, \forall x \in [a,b]$ (g ist konstant):

Dann ist $0 = g^\prime(x) = \frac{f(b) - f(a)}{b-a} - f^\prime(x)$, und jeder Punkt $x \in (a,b)$ erfüllt die behauptete Gleichung.

2. Fall $g$ ist nicht Konstant auf $[a,b]$:

$[a,b] \subseteq \mathbb{R}$ kompakt ist $g$ in $[a,b]$ stetig, gibt es $x_{min}, x_{max} \in [a,b]$ so, dass 
$g(x_{min} = \inf\{g(x) \vert x \in [a,b]\} < \sup\{g(x) \vert x \in [a,b]\} = x_{max}$.

Außerdem ist 

$$g(a) = \frac{f(b) - f(a)}{b-a} \cdot a - f(a) = 
\frac{f(b) \cdot a \cancel{- f(a) \cdot a} - f(b) \cdot a \cancel{+ f(a) \cdot a}}{b-a} =
$$

$$
\frac{f(b) \cdot b  - f(a) \cdot b - f(b)(b-a)}{b-a} =
\frac{f(b) - f(a)}{b-a} \cdot b - f(b) = g(b)
$$

Ist bspw. $x_{min} \in (a,b)$ und $x_{max} \in (a,b)$, so gilt nach Satz VII.7:

$$0 = g^\prime(x_{min}) = \frac{f(b) - f(a)}{b-a} - f^\prime(x_{x_min})$$

Die Behauptung folgt nun mit $x_0 := x_{min} \in (a,b)$.


## VII.3 Höhere Ableitungen und der Satz von Taylor $\Leftarrow$ Wichtig

### Definition VII.11 

Seien $U \subseteq \mathbb{R}, U \neq \emptyset$, offen, $f: U \rightarrow \mathbb{R}$ und $n \in \mathbb{N}$.

- (i) $f$ bei $x_0 \in U$ n-mal diff'bar $:\Leftrightarrow$ Es gibt ein $\delta > 0$ so, dass für alle $x \in (x_0 - \delta, x_0 + \delta)$ 

$f$ ist diff'bar bei $x$ und $f^\prime(x) := \frac{df}{dx}(x)$,

$f^{(1)} := f^\prime$ ist diff'bar bei $x$ und $f^{\prime\prime}(x) = \frac{d}{dx}[f^\prime](x)$,

$f^{(2)} := f^{\prime\prime}$ ist diff'bar bei $x$ und $f^{\prime\prime\prime}(x) = \frac{d}{dx}[f^{\prime\prime}](x)$,

$\quad\quad\quad\quad\quad\quad\quad\quad\quad\vdots$

$f^{(n-2)}$ ist diff'bar bei $x$ und $f^{(n-1)}(x) = \frac{d}{dx}[f^{(n-2)}](x)$,

$f^{(n-1)}$ ist diff'bar bei $x_0$ und $f^{(n)}(x_0) = \frac{d}{dx}[f^{(n)}](x_0)$.

- (ii) $f$ heißt auf $U$ n-mal diff'bar $:\Leftrightarrow \forall x_0 \in U: f$ ist bei $x_0$ n-mal diff'bar.

## Satz VII.12 [Satz von Taylor]:

Seien $a,b,x_0 |in \mathbb{R}$ mit $a < x_0 < b, n \in \mathbb{N}$ und $f: [a,b] \rightarrow \mathbb{R}$ auf $(a,b)$ $(n+1)$-mal diff'bar.

Dann gibt es zu jedem $x \in (a,b) \{x_0\}$ ein $x^\prime \in (x, x_0)$ [falls $x < x_0$] bzw. $x^\prime \in (x, x_0)$ [falls $x > x_0$] so, dass 

$$f(x) = T_n[f, x_0; x] + R_{n+1} [f, x_0; x]$$

wobei 

$$T_n[f, x_0; x] := \sum\limits^n_{k=0} \frac{(x-x_0)^k}{k!} f^{(k)}(x_0) = f(x_0) + (x-x_0)f^\prime(x_0 + \frac{(x-x_0)}{2} \cdot f^{\prime\prime}(x_0) + \dots + \frac{(x-x_0)^n}{n!} \cdot f^{(n)}(x_0),$$

das Taylorpolynom $n$.Ordnung ($T_n$) und

$$R_{n+1} [f, x_0; x] = \frac{(x-x_0)^{n+1}}{(n+1)!} f^{n+1}(x^\prime)$$

das Taylorsche Restglied $(n+1)$.Ordnung sind.

Beispiel:

Sei $f(x) = \sin(x), (a,b) = (-\infty, \infty), x_0 = 0$,

$$
f^{0}(x) = \sin(x), f^{1}(x) = \cos(x), f^{2}(x) = -\sin(x), f^{3}(x) = -\cos(x), f^{4}(x) = \sin(x), \dots
$$

$n=3$:

$$T_3[\sin, 0; x] = \sin(0) + x \cdot \cos(0) + \frac{x^2}{2}\cdot(-\sin(0)) + \frac{x^3}{6}\cdot(-\cos(0)) = x - \frac{x^3}{6}$$

$$R_4 [\sin, 0; x] = \frac{x^4}{24}$$

$$\Rightarrow \forall x \in \mathbb{R}:\quad \left|sin(x) - (x-\frac{x^3}{6})\right| \leq \frac{x^4}{24}$$

$$x = \frac{1}{10} \Rightarrow \left|sin(\frac{1}{10}) - (\frac{1}{10}-\frac{1}{6000})\right| \leq \frac{1}{240.000}$$
